```python
d1 = {
    'Key1' : [1,2,3,4,5,6],
    'Key2' : [7,8,9,10,11,12],
    'Key3' : [13,14,15,16,17,18],
    'Key4' :  [19,20,21,22,23,24]
}
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(d1)
```

```python
pd.DataFrame(d1, index=['a','b','c','d','e','f'])
```

```python
d2 = {
    'Key1' : [1,2,3,'Ram',5,6],
    'Key2' : [7,8,9,10,11,12],
    'Key3' : [13,14,'Suresh',16,17,18],
    'Key4' :  [19,'Lakshmi',21,22,23,24]
}
```

```python
pd.DataFrame(d2, index=['a','b','c','d','e','f'])
```

```python
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how= 'inner', on = 'Key1')
```

```python
pd.merge(df1,df2, how= 'left', on = 'Key4')
```

```python
pd.merge(df1,df2, how= 'outer', on = 'Key1')
```
